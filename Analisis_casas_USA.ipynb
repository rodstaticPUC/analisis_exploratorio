# Relacion entre variables discretas

# Matriz de correlacion

```python
cormat = df_train.corr()
```

```python
cormat # Id versus ID, muestra una correlacion perfecta (1), pero es la misma varibale
```

```python
f, ax = plt.subplots(figsize=(12,9)) 
sns.heatmap(cormat, vmax=0.8, square=True) 

# Mientras mas claro son los cuadrados mayor correlacion existe.
```

Observando el heat map, SalePrice tiene una gran correlacion con OverallQual, GrLivArea, GarageYrBlt

```python
k = 10
```

```python
cols = cormat.nlargest(k, 'SalePrice')['SalePrice'].index # identificamos las 10 variables mas correlacionadas con SalePrice
```

```python
cols
```

```python
cm = np.corrcoef(df_train[cols].values.T)
```

```python
cm #arreglo de correlacion traspuesto
```

```python
sns.set(font_scale=1.25) # escale de la fuente sea 1.25 de tamaño
```

```python
hm=sns.heatmap(cm,cbar=True, annot=True, square = True, fmt ='.2f', annot_kws= {'size':10}, yticklabels=cols.values, xticklabels=cols.values)
```

# Conclusiones

Se aprecia que la calidad (0.79), el area construida (0.71), cantidad de garages (0.64), area del garage (0.62), si tiene sotano (0.61), area del primer piso (0.61), si tiene baño completo (0.56), total de habitaciones sobre el sotano (0.53) y finalmente el año de construccion de la casa (0.52).

Tambien se aprecia una colinielidad entre el area del garage y la cantidad de auto que puede soportar el garage, ya que ambas variables estan relacionadas estrechamente (a mayor area de garage, mayor cantidad de autos en el garage).
